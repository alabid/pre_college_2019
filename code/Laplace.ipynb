```python
##
## How epsilon affects accuracy of sum estimates
##

import numpy as np
import laplace
import matplotlib.pyplot as plt

n = 1000
trials = 50
epsilons = 0.1*np.arange(1, 21) # epsilons = [0.1, 0.2, ... 2.0]

diffs = []
dataset = np.random.random(n)
for eps in epsilons:
    a = laplace.non_private_sum(dataset)
    diffs_per_epsilon = []
    for t in range(trials):
        b = laplace.private_sum(dataset, n, eps)
        diffs_per_epsilon.append(abs(a - b))
    diffs.append(np.mean(diffs_per_epsilon))
    
plt.plot(epsilons, diffs)
plt.title("Effect of Epsilon on Absolute Error for Sums")
plt.xlabel("Absolute Error")
plt.ylabel("Epsilon")
plt.show()
```

```python
##
## How epsilon affects accuracy of mean estimates
##

import numpy as np
import laplace
import matplotlib.pyplot as plt

n = 1000
trials = 50
epsilons = 0.1*np.arange(1, 21) # epsilons = [0.1, 0.2, ... 2.0]

diffs = []
dataset = np.random.random(n)
for eps in epsilons:
    a = laplace.non_private_mean(dataset)
    diffs_per_epsilon = []
    for t in range(trials):
        b = laplace.private_mean(dataset, n, eps)
        diffs_per_epsilon.append(abs(a - b))
    diffs.append(np.mean(diffs_per_epsilon))
    
plt.plot(epsilons, diffs)
plt.title("Effect of Epsilon on Absolute Error for Averages")
plt.xlabel("Absolute Error")
plt.ylabel("Epsilon")
plt.show()
```
